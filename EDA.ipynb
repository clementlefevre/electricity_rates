```python
df = pd.read_stata('data/tempForClement.dta')
df.to_csv('data/dataset1.csv', index=None)
```

```python
def select_munid(df,munid):
    df = df[df['munid']==munid]
    df.drop(['l2','l3'],axis=1,inplace=True)
    df['mean_comsumption_kWh']=df.sales*1.0/df.user
    return df
```
